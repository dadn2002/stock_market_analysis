# **Utilizando yfinance para ações brasileiras**

A biblioteca possui basicamente três módulos:

* yf.Ticker
* yf.download
* yf.pandas_datareader

Quase todos os métodos estão no módulo yf.Tickers

O módulo **download** serve para baixar rapidamente os dados históricos de vários tickers de uma só vez.



# **Como baixar dados de ativos?**


```python
import yfinance as yf
```

```python
petro = yf.Ticker('PETR4.SA')
```

```python
petro.ticker
```

```python
petro.actions
```

```python
petro.dividends
```

```python
petro.splits
```

# **Obtendo agora dados de cotação**

```python
petro.history(start = '2023-01-23', end  = '2023-10-01', interval ='1m')
```

```python
petro.history(start = '2023-10-05', end  = '2023-10-09', interval ='1m')
```

```python
petro.history(start = '2023-10-05', end  = '2023-10-09', interval ='5m')
```

```python
petro.history(period = 'max', interval ='1wk')
```

# **Extração de cotação multiticker**

```python
yf.download("WEGE3.SA", start = '2023-01-01', end = '2023-10-01')
```

```python
yf.download(["WEGE3.SA", "IVVB11.SA"], start = '2023-01-01', end = '2023-10-01')
```

```python
yf.download("WEGE3.SA IVVB11.SA VALE3.SA", start = '2023-01-01', end = '2023-10-01', group_by = 'column')
```

```python
yf.download("WEGE3.SA IVVB11.SA VALE3.SA", start = '2023-01-01', end = '2023-10-01', group_by = 'tickers')
```
